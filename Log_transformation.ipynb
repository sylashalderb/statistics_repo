```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox

# Generate right-skewed data
np.random.seed(42)
skewed_data = np.random.exponential(scale=2.0, size=5000)  # Exponentially distributed data

# Apply a logarithmic transformation to reduce skewness
transformed_data = np.log(skewed_data + 1)

# Plot the original skewed data
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.hist(skewed_data, bins=50, density=True, alpha=0.75, color='blue')
plt.title("Original Skewed Data")
plt.xlabel("x")
plt.ylabel("Density")

# Plot the transformed data
plt.subplot(1, 3, 2)
plt.hist(transformed_data, bins=50, density=True, alpha=0.75, color='orange')
plt.title("Log-Transformed Data")
plt.xlabel("x (log)")
plt.ylabel("Density")

plt.show()
```
