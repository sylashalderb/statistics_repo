```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 2.5
x_values = np.array([1, 2, 3, 4, 5])

# Compute P(x)
p_values = x_values ** -alpha

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x_values, p_values, marker='o', linestyle='-', color='b', label=r'$P(x) = x^{-\alpha}$')  # Note the r before the string
plt.title('Plot of $P(x) = x^{-\\alpha}$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('P(x)', fontsize=12)
plt.xticks(x_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 2.5
x_values = np.array([1, 2, 3, 4, 5])

# Compute unnormalized P(x)
p_values = x_values ** -alpha

# Compute normalization factor Z
Z = np.sum(p_values)

# Compute normalized P(x)
p_values_normalized = p_values / Z

# Plot unnormalized and normalized probabilities
plt.figure(figsize=(10, 6))
plt.plot(x_values, p_values, marker='o', linestyle='-', color='b', label=r'Unnormalized $P(x) = x^{-\alpha}$')  # Corrected string
plt.plot(x_values, p_values_normalized, marker='o', linestyle='--', color='r', label=r'Normalized $P(x)$')  # Corrected string
plt.title('Unnormalized vs Normalized $P(x)$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('$P(x)$', fontsize=12)
plt.xticks(x_values)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.show()
```
