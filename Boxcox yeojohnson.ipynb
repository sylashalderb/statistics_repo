```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generating a sample data (Pareto Distribution in this case)
alpha = 3
x_min = 2
x = np.arange(x_min, 11)

# Pareto distribution formula: P(x) = (alpha * x_min^alpha) / (x^(alpha+1))
P_x = (alpha * x_min**alpha) / (x**(alpha + 1))

# Box-Cox transformation (with lambda=0.5)
P_x_boxcox, lambda_boxcox = stats.boxcox(P_x)

# Yeojohnson transformation (with lambda=0.5)
P_x_yeojohnson, lambda_yeojohnson = stats.yeojohnson(P_x)

# Plotting the results
plt.figure(figsize=(10, 6))

# Original Pareto data
plt.subplot(1, 3, 1)
plt.plot(x, P_x, label='Original Data', color='blue')
plt.title('Original Data (Pareto Distribution)')
plt.xlabel('X')
plt.ylabel('P(x)')
plt.grid(True)

# Box-Cox transformed data
plt.subplot(1, 3, 2)
plt.plot(x, P_x_boxcox, label='Box-Cox Transformed', color='green')
plt.title(f'Box-Cox Transformed (={lambda_boxcox:.2f})')
plt.xlabel('X')
plt.ylabel('Transformed P(x)')
plt.grid(True)

# Yeojohnson transformed data
plt.subplot(1, 3, 3)
plt.plot(x, P_x_yeojohnson, label='Yeojohnson Transformed', color='red')
plt.title(f'Yeojohnson Transformed (={lambda_yeojohnson:.2f})')
plt.xlabel('X')
plt.ylabel('Transformed P(x)')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
